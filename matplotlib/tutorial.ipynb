```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Creating BAr CHART

```python
x =  ["PYTHON", "c++", "C", "JAVA"]
y =  [50,80,40,60]
z = [20,10,20,30]
plt.xlabel("language",fontsize=15)
plt.ylabel("NO",fontsize=15)
plt.title("VSCODE",fontsize=15)
# c = ["y","b","g","r"]
# plt.bar(x,y,width=0.2,color=c, align = "center", edgecolor = "orange", linewidth =4, linestyle =":")    # ye line se apan bar ki size aur coloring dete hai manual  colors bhi de sakte hai bas kisi ek char me store karake color ko jaha par properties hai udhar call kar do
plt.bar(x,y,width=0.2,color="red", align = "center", label="popularity")
plt.bar(x,z,width=0.2,color="yellow", align = "center", label="popularity1")
plt.legend()   
plt.show()
```

```python
x =  ["PYTHON", "c++", "C", "JAVA"]
y =  [50,80,40,60]
z = [20,10,20,30]


width=0.2
p = np.arange(len(x))   ## abhi x axis me jitne bhi elements hai vo index wise arranges hai jaise python 0 pe , c++ hai 1 pe aur bakki sab respective place pe
p1 = [j+width for j in p]

plt.xlabel("language",fontsize=15)
plt.ylabel("NO",fontsize=15)
plt.title("VSCODE",fontsize=15)

plt.bar(p,y,width,color="red", align = "center", label="popularity")
plt.bar(p1,z,width,color="yellow", align = "center", label="popularity1")

plt.xticks(p+width/2, x,rotation= 20)
plt.legend()   
plt.show()
```

### horizontal bar

```python
x =  ["PYTHON", "c++", "C", "JAVA"]
y =  [50,80,40,60]
z = [20,10,20,30]

plt.xlabel("language",fontsize=15)
plt.ylabel("NO",fontsize=15)
plt.title("VSCODE",fontsize=15)

plt.barh(x,y,color="red", align = "center", label="popularity")    # bar ki jagah barh karna hai jisme h ki wajah se horizontal ho jayega graph
plt.barh(x,z,color="yellow", align = "center", label="popularity1")

plt.legend()   
plt.show()
```

### SCATTERPLOT

```python
x =[1,2,3,4,5,6,7,8,9]
y =[2,4,6,8,1,3,5,7,9]
colors = ["r","b","g","y","r","y","r","b","g"]
sizes = [200,400,600,800,900,100,200,300,400]
# plt.scatter(x,y,c= colors, s= sizes, edgecolors="g", marker="*", linewidths=2)
plt.scatter(x,y,c= colors, s= sizes, cmap="viridis")
t=plt.colorbar()
t.set_label ("colors", fontsize = 20)
plt.title("Scatter plot", fontsize = 30)
plt.xlabel("xaxis", fontsize = 20)
plt.ylabel("yaxis", fontsize = 20)

plt.show()
```

### piechart graph


```python
x = [10,20,30,40]
y = ["c","c++", "python", "java"]
ex = [0,0.2,0,0]
c=["r","b","g","y"]
plt.pie(x, labels=y, explode=ex,colors=c, autopct="%0.1f%%",shadow="True",radius=1.5,labeldistance=1.1,
        startangle=0,textprops={"fontsize":15},counterclock=False,wedgeprops={"edgecolor":"r"},rotatelabels="True")
plt.show()
```
