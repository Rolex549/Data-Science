```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# DATA STRUCTURES : [RESTAURANT_ID, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],   #paradise biryani
    [2, 120000, 140000, 160000, 190000],   # Beijing Bites
    [3, 200000, 230000, 260000, 300000],   #Pizza Hub
    [4, 180000, 210000, 240000, 270000],   #Burger Point
    [5, 160000, 185000, 205000, 230000]    #Chai Point
])

print("======Zomato Sales Data Analysis======")
print("sales data shape: ", sales_data.shape)
print("\n sample data for first 3 restaurants: ", sales_data[:,3:])
```

#### TOTAL SALES PER YEAR

```python
# method 1
print(np.sum(sales_data, axis = 0))


#method 2
yearly_sales = np.sum(sales_data[:,1:], axis = 0)
print("yearly sales from 2021 to 2024: \n", yearly_sales)


#### minimum sales per restaurant
min_sales = np.min(sales_data[:,1:], axis=1)
print("minimum sales per restaurant: \n", min_sales)


#### maximum sales per restaurant
max_sales = np.max(sales_data[:, 1:,], axis=0)
print("maximum sales: \n", max_sales)

###AVERAGE SALES PER YEAR
avg_sales = np.mean(sales_data[:,1:], axis=0)
print("average sales per year: \n", avg_sales)


#### cumulative sales per year
cum_sales = np.cumsum(sales_data[:,1:], axis=0)
print("cumulative sales per year: \n", cum_sales)
```

#### GRAPH PLOT KAISE KARTE HAI


```python
plt.figure(figsize=(15,6))
plt.plot(np.mean(cum_sales, axis=0))
plt.title("Average cumulative sales across all restaurant")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
```

#### VECTOR ADDITION AND MULTIPLICATION and DOT PRODUCT and ANGLES BETWEEN VECTORS

```python
vector1 = np.array([1,2,3])
vector2 = np.array([4,5,6])
print("vector addition: \n", vector1 + vector2)
print("vector multiplication: \n", vector1 * vector2)
print("dot product: \n", np.dot(vector1, vector2)) 
angle = np.arccos(np.dot(vector1, vector2)/ (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print("angle between two vectors in radians: \n", angle)
print("angle between two vectors in degrees: \n", np.degrees(angle))
```
