### CONVERSION OF ANY PICTURE INTO DARK MODE

```python
import imageio.v3 as iio
import matplotlib.pyplot as plt
import numpy as np

try:
    logo = iio.imread('pandas.png')

    print("Shape:", logo.shape, "Min:", logo.min(), "Max:", logo.max(), "Dtype:", logo.dtype)

    logo_normalized = logo.astype(float) / 255.0

    # If the image has an alpha channel (4th channel), separate it
    if logo_normalized.shape[-1] == 4:
        rgb = logo_normalized[..., :3]
        alpha = logo_normalized[..., 3:]
    else:
        rgb = logo_normalized
        alpha = None

    # Invert only the RGB channels
    dark_rgb = 1 - rgb

    # Reattach alpha channel if present
    if alpha is not None:
        darklogo = np.concatenate([dark_rgb, alpha], axis=-1)
    else:
        darklogo = dark_rgb

    # Display both
    plt.figure(figsize=(10,5))
    plt.subplot(121)
    plt.imshow(logo_normalized)
    plt.title("Original Image")
    plt.axis('off')
   

    plt.subplot(122)
    plt.imshow(darklogo)
    plt.title("Dark Mode Image")
    plt.axis('off')
    plt.show()

except FileNotFoundError:
    print("File not found. Please ensure the file path is correct.")
```
