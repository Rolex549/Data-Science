#  IMPORTING LIBRARIES

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

# USE SAMPPLE DATA

```python
data = {
    'LotArea': [8500, 9600, 11250, 9550, 14260],
    'GrLivArea': [1710, 1262, 1786, 1717, 2198],
    'OverallQual': [7, 6, 8, 7, 9]
}
```

# DATASET KO DATAFRAME ME DAALO

```python
df = pd.DataFrame(data)
print("ðŸ”¹ Original Data:\n", df)
```

# Standard scaling

```python
std_scaler = StandardScaler()
df_std = pd.DataFrame(std_scaler.fit_transform(df), columns=df.columns)
print("\nðŸ”¹ Standard Scaled Data (mean=0, std=1):\n", df_std)
```

# Min-Max Scaling (0 to 1)


```python
minmax_scaler = MinMaxScaler()
df_minmax = pd.DataFrame(minmax_scaler.fit_transform(df), columns=df.columns)
print("\nðŸ”¹ Min-Max Scaled Data (range 0-1):\n", df_minmax)
```
