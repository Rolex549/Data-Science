# IMPORTING LIBRARIES


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

# CREATE A SAMPLE DATASET

```python
data = {
    'City': ['Delhi', 'Mumbai', 'Kolkata', 'Delhi', 'Mumbai', 'Kolkata'],
    'HouseStyle': ['Bungalow', 'Apartment', 'Villa', 'Villa', 'Bungalow', 'Apartment'],
    'Price': [75, 55, 90, 85, 70, 60]
}
```

# CONVERT IT INTO DATAFRAME

```python
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
```

# LABEL ENCODING

```python
label_encoder = LabelEncoder()
df['HouseStyle_Label'] = label_encoder.fit_transform(df['HouseStyle'])
print(df)
```

# ONE HOT ENCODING

```python
df_encoded = pd.get_dummies(df, columns=['City'])

print("\nEncoded DataFrame:")
print(df_encoded)
```

# VISUALIZATION

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='HouseStyle_Label', y='Price', data=df_encoded)
plt.title('House Price vs Encoded HouseStyle')
plt.xlabel('HouseStyle (Label Encoded)')
plt.ylabel('Price (in Lakhs)')
plt.xticks(ticks=[0, 1, 2], labels=label_encoder.classes_)
plt.grid(True)
plt.show()
```
