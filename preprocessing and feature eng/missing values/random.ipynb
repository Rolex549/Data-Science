```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv')
```

# Mean and Median Imputation


```python

df_mean = df.copy()
df_mean['Age'].fillna(df_mean['Age'].mean(), inplace=True)

df_median = df.copy()
df_median['Age'].fillna(df_median['Age'].median(), inplace=True)

```

# Random Imputation


```python
df_random = df.copy()
non_null = df_random['Age'].dropna()
imputed = np.random.choice(non_null, size=df_random['Age'].isnull().sum(), replace=True)
df_random.loc[df_random['Age'].isnull(), 'Age'] = imputed
```

# KDE Plot



```python
sns.kdeplot(df_mean['Age'], label='Mean Imputation')
sns.kdeplot(df_median['Age'], label='Median Imputation')
sns.kdeplot(df_random['Age'], label='Random Imputation')
plt.title('Age Distribution Comparison')
plt.legend()
plt.show()
```

```python
df_viz = pd.DataFrame({
    'Mean': df_mean['Age'],
    'Median': df_median['Age'],
    'Random': df_random['Age']
})
sns.violinplot(data=df_viz)
plt.title('Violin Plot of Age Imputation Methods')
plt.show()
```
