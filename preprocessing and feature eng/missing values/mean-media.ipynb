```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

# DAtaset

```python
data = {
    'Age': [25, np.nan, 30, 22, np.nan, 28],
    'Salary': [50000, 60000, np.nan, 52000, 58000, np.nan]
}
df = pd.DataFrame(data)
print("Original Data:\n", df)
```

# Mean imputation

```python
mean_imputer = SimpleImputer(strategy='mean')
df_mean = df.copy()
df_mean[['Age', 'Salary']] = mean_imputer.fit_transform(df_mean[['Age', 'Salary']])
print("\nAfter Mean Imputation:\n", df_mean)
```

# Median Imputation


```python
median_imputer = SimpleImputer(strategy='median')
df_median = df.copy()
df_median[['Age', 'Salary']] = median_imputer.fit_transform(df_median[['Age', 'Salary']])
print("\nAfter Median Imputation:\n", df_median)
```

# Plotting histograms


```python
plt.subplot(2, 2, 1)
plt.hist(df_mean['Age'], bins=5, color='skyblue', edgecolor='black')
plt.title('Age - Mean Imputation')

plt.subplot(2, 2, 2)
plt.hist(df_median['Age'], bins=5, color='lightgreen', edgecolor='black')
plt.title('Age - Median Imputation')
```

```python
plt.subplot(2, 2, 3)
plt.hist(df_mean['Salary'], bins=5, color='salmon', edgecolor='black')
plt.title('Salary - Mean Imputation')

plt.subplot(2, 2, 4)
plt.hist(df_median['Salary'], bins=5, color='orange', edgecolor='black')
plt.title('Salary - Median Imputation')

plt.tight_layout()
plt.show()
```
