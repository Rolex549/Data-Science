```python
from sklearn.impute import KNNImputer
import pandas as pd
import matplotlib.pyplot as plt

```

```python
data = pd.DataFrame({
    'Pclass': [1, 2, 3, 1, 2],
    'Age': [22, None, 24, 35, None],
    'Fare': [71.28, 10.5, 8.05, 53.1, 12.3]
})
```

```python
df = pd.DataFrame(data)
```

```python
imputer = KNNImputer(n_neighbors=2)
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)
imputed_age = df_imputed['Age']
```

```python
original_age = df['Age'].dropna()
```

```python
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.hist(original_age, bins=5, color='skyblue', edgecolor='black')
plt.title('Original Age (before imputation)')
plt.xlabel('Age')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
plt.hist(imputed_age, bins=5, color='lightgreen', edgecolor='black')
plt.title('Age after KNN Imputation')
plt.xlabel('Age')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```
