```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic.csv')
```

```python
age_original = df['Age'].dropna()
```

```python
mean_age = df['Age'].mean()
std_age = df['Age'].std()
```

#  Define tail value (mean + 3 * std)


```python
tail_value = mean_age + 3 * std_age
```

# Impute missing values in 'age' column


```python
df['age_imputed'] = df['Age'].fillna(tail_value)
```

```python
plt.figure(figsize=(12, 5))
```

# Original age distribution


```python
plt.subplot(1, 2, 1)
plt.hist(age_original, bins=30, color='skyblue', edgecolor='black')
plt.title('Original Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
```

# Imputed age distribution


```python
plt.subplot(1, 2, 2)
plt.hist(df['age_imputed'], bins=30, color='salmon', edgecolor='black')
plt.title('Imputed Age Distribution (End-of-Distribution)')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
